Il Linguaggio dei Furfanti

In Svezia, i bambini giocano spesso utilizzando un linguaggio un po' particolare detto rövarspråket, che significa "linguaggio dei furfanti": consiste nel raddoppiare ogni consonante di una parola e inserire una "o" nel mezzo. Ad esempio la parola "mangiare" diventa "momanongogiarore".

Scrivi una funzione in grado di tradurre una parola o frase passata tramite input in rövarspråket. Dopo aver tradotto una frase, il programma dovrà chiedere all'utente se intende tradurne un'altra, e in caso di risposta positiva, dovrà attendere l'inserimento di una nuova parola da parte dell'utente.

```python
from random import random


def traduttore():
    print("Ciao! questo programma traduce un testo in ""rövarspråket"
          "")
    vocali = "aeiou"
    while True:
        text = input("Inserisci il testo che desideri tradurre: ")
        translation = ""
        for letter in text:
            if letter in vocali:
                translation += letter
            else:
                translation = translation + letter + "o" + letter

        print("Ecco a te la traduzione" + translation)

        if input("Desidere tradurre un'altra frase? ") == "no":
            break
```

Scrivi una funzione a cui viene passata una parola e riconosce se si tratta di un palindromo (parole che si leggono uguali anche al contrario) oppure no.

```python
def inversa():
    phrase = str(input("Inserisci la frase: "))
    phraseReverce = phrase[::-1]
    if phraseReverce == phrase:
        print("è un palindromo")
    else:
        print("non è un palindromo")


inversa()
#giusto ma nella correzione è in modo diverso
```

Scrivi una funzione generatrice di password.

La funzione deve generare una stringa alfanumerica di 8 caratteri qualora l'utente voglia una password semplice, e di 20 caratteri ASCII qualora desideri una password più complicata.

```python
import random
import string


def password():
    choice = str(input("vuoi una password sicura? yes/no : "))
    ascii = string.digits + string.ascii_letters + string.punctuation
    noAscii = string.digits + string.ascii_letters

    if choice == "yes":
        lunghezza = 20
        type = ascii
    else:
        lunghezza = 8
        type = noAscii

    psw = ""
    counter = 0

    while counter < lunghezza:
        char = random.choice(type)
        psw += char
        counter += 1

    print("la tua password è: " + psw)


password()
```

Scrivi una funzione che calcoli la somma (espressa in MB) delle dimensioni dei file presenti nella cartella di lavoro utilizzando il modulo os.

```python
import os


def file_size():
    total = 0
    files = os.getcwd()
    for file in os.listdir(files):
        total += os.path.getsize(os.path.join(files, file))

    print(f"La somma delle dimensioni dei file presenti nella cartella '{files}' è: {(total / 1048576):.2f}MB")


file_size()
```

Scrivi una funzione "postino" che sia in grado di spedire delle eMail tramite Gmail.

Suggerimento: puoi usare il modulo smtplib.

```python
import smtplib


#studiare funzioni di smtplib
def postino():
    username = input("Inserisci il tuo username: ")
    password = input("Inserisci la tua password: ")
    destinatario = input("Inserisci l'email del destinatario: ")
    oggetto = input("Inserisci l'oggetto della mail: ")
    messaggio = input("Ora puoi inserire il messaggio che vuoi inviare: ")
    contenuto = f"Subject: {oggetto}\n\n{messaggio}"
    print("Sto effettuando la connessione col Server...")
    email = smtplib.SMTP("smtp.gmail.com", 587)
    email.ehlo()  #connessione al server
    email.starttls()
    email.login(username, password)
    print("Sto inviando...")
    email.sendmail(username, destinatario, contenuto)
    email.quit()
    print("Messaggio Inviato!")


postino()
```

Scrivi una funzione "cercatrice" che scansioni un dato percorso di sistema alla ricerca di file di tipo pdf tramite il modulo os. La funzione dovrà avere le seguenti caratteristiche:

Il percorso fornito dovrà essere anzitutto validato, in quanto deve portare a una cartella esistente
La funzione dovrà fornire un elenco dei file pdf (con/relativo/percorso) man mano che questi vengono trovati
In fine la funzione dovrà fornire in output il totale dei file .pdf che sono stati trovati durante la scansione.

```python
import os


def cercatrice(percorso):
    if not os.path.isdir(percorso):
        print(f"Il percorso inserito '{percorso}' risulta non essere un percorso idoneo. Verifica e riprova, grazie.")
        return None
    contatore = 0
    print(f"Sto effettuando la scansione di '{percorso}' alla ricerca di file .pdf")
    for cartella, sottocartelle, files in os.walk(percorso):
        for file in files:
            if file.endswith(".pdf"):
                pdf = os.path.join(cartella, file)
                print(f"Trovato file pdf: {pdf}")
                contatore += 1
    print(f"Scansione Ultimata! Ho trovato {contatore} files con estensione pdf.")
```

```python
def fattoriale(n):
    if n == 1:
        return n
    else:
        result = n * fattoriale(n - 1)
        return result
```

```python
def fibonacci(n):
    if n <= 1:
        return n
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)


limite = int(input("Inserisci il numero di valori della serie che desideri vedere: "))

for num in range(1, limite + 1):
    print(fibonacci(num))

```

```python
stringa = str(input("inserisci tringa da modificare"))
key = 2


def cifrario_cesare(stringa, key):
    alfabeto = 'abcdefghijklmnopqrstuvwxyz'

    stringa_modificata = ''

    for carattere in stringa:
        if carattere.lower() in alfabeto:

            index = alfabeto.index(carattere.lower())

            nuovo_indice = (index + key) % 26

            stringa_modificata += alfabeto[nuovo_indice] if carattere.islower() else alfabeto[nuovo_indice].upper()
        else:
            stringa_modificata += carattere

    return stringa_modificata


cifrario_cesare(stringa, key)
```
